```python
'''
Rock, Paper, Scissors

'''
import random

def getWeights(x,y,z):
    '''
    Default weights for Rock, Paper, Scissors are 1 each when x, y, z are 0
    Every time a user plays Rock, the weight for Paper is incremented by 1, and so on and so forth for the other options
    '''
    return (z+1,x+1,y+1)

comp_choices = ['Rock','Paper','Scissors']
user_choices = {'1':'Rock', '2':'Paper', '3':'Scissors', '4':'Quit Game'}
exit_loop_check = ''
score_user = 0
score_comp = 0
loop_counter = 0
rock_counter = 0
paper_counter = 0
scissors_counter = 0
print("Rock, Paper, Scissors\n")
while exit_loop_check != 'Quit Game':
    weights_for_comp_choice = getWeights(rock_counter,paper_counter,scissors_counter)
    curr_comp_choice = random.choices(comp_choices, weights = weights_for_comp_choice)[0]
    curr_user_choice = input("Enter the number associated with your choice:\n1. Rock\n2. Paper\n3. Scissors\n4. Quit Game\n")
    if curr_user_choice in user_choices:
        exit_loop_check = user_choices[curr_user_choice]
        if user_choices[curr_user_choice] != 'Quit Game':
            loop_counter += 1
            print("Your Choice: ",user_choices[curr_user_choice])
            print("Computer's Choice: ",curr_comp_choice,"\n")
            if user_choices[curr_user_choice] == 'Rock':
                rock_counter += 1
                if user_choices[curr_user_choice] == curr_comp_choice:
                    print("WHOA! Thats a draw\nYou ",score_user,"-",score_comp," Computer")
                elif curr_comp_choice == 'Paper':
                    score_comp += 1
                    print("OH NO! You have lost\nYou ",score_user,"-",score_comp," Computer")
                else:
                    score_user += 1
                    print("HURRAH! You have won\nYou ",score_user,"-",score_comp," Computer")
            elif user_choices[curr_user_choice] == 'Paper':
                paper_counter += 1
                if user_choices[curr_user_choice] == curr_comp_choice:
                    print("WHOA! Thats a draw\nYou ",score_user,"-",score_comp," Computer")
                elif curr_comp_choice == 'Scissors':
                    score_comp += 1
                    print("OH NO! You have lost\nYou ",score_user,"-",score_comp," Computer")
                else:
                    score_user += 1
                    print("HURRAH! You have won\nYou ",score_user,"-",score_comp," Computer")
            elif user_choices[curr_user_choice] == 'Scissors':
                scissors_counter += 1
                if user_choices[curr_user_choice] == curr_comp_choice:
                    print("WHOA! Thats a draw\nYou ",score_user,"-",score_comp," Computer")
                elif curr_comp_choice == 'Rock':
                    score_comp += 1
                    print("OH NO! You have lost\nYou ",score_user,"-",score_comp," Computer")
                else:
                    score_user += 1
                    print("HURRAH! You have won\nYou ",score_user,"-",score_comp," Computer")
    else:
        print("ERROR: Invalid Input")
print("Thanks for playing. You played ",loop_counter," game(s) out of which you won ",score_user," game(s).")
```
