```python
'''
Voters in Florida - Text Extraction
'''

import re

fp = open("FloridaVoters.html", "r")
subset_of_text = str(fp.read()).split("<h1>Voter Registration - Current by County</h1>")
fp.close()
items = subset_of_text[1].split("<tr>\n<td>")
county_list = []
rep_total = 0
dem_total = 0
for i in range(len(items)):
    if i!=0 and i!=len(items)-1:
        county = re.findall("[^<>\/\\n (tr)(td)]+",items[i])
        if len(re.findall("[A-Za-z]+",county[1][0]))>0:
            county[2] = int(county[2].replace(",",""))
            county[3] = int(county[3].replace(",",""))
            county_tuple = (county[0]+" "+county[1],county[2],county[3])
            rep_total += county[2]
            dem_total += county[3]
        else:
            county[1] = int(county[1].replace(",",""))
            county[2] = int(county[2].replace(",",""))
            county_tuple = (county[0],county[1],county[2])
            rep_total += county[1]
            dem_total += county[2]
        county_list = county_list + [county_tuple]
sorted_list = sorted(county_list, key = lambda county_tuple:county_tuple[2])
for line in sorted_list:
    print(line[0],line[1],line[2])
print("Total",rep_total,dem_total)
```
